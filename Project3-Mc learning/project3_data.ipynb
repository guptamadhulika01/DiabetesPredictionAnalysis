```python
# read county file to DF 

County_population_df = pd.read_csv('county_population.csv', delimiter=',')
County_population_df.head()
```

```python
County_Population_df_2016 = County_population_df[['State','County','Population Estimate, 2012','Population Estimate, 2016']]
County_Population_df_2016.head()
```

```python
County_Population_df_2016[["Population Estimate, 2012", "Population Estimate, 2016"]] = County_Population_df_2016[["Population Estimate, 2012", "Population Estimate, 2016"]].apply(pd.to_numeric)
```

```python
# to take out percentage change in population from 2012 - 2016
County_Population_df_2016['PercChangeInPop2012-2016']=County_Population_df_2016.apply(lambda row: ((row['Population Estimate, 2016']- row['Population Estimate, 2012']) /row['Population Estimate, 2012'] )* 100, axis=1)
```

```python
#df1['score_rounded_off_single_decimal']= round(df1['Score'],1)

County_Population_df_2016['PercChangeInPop2012-2016']=round(County_Population_df_2016['PercChangeInPop2012-2016'],2)
County_Population_df_2016.head()
```

```python
#save to csv
County_Population_df_2016.to_csv('./FeatureDataFiles/County_Population_df_2016.csv')
```
